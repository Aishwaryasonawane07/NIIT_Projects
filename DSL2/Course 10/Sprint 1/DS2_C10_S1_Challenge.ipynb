```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = le.fit_transform(df[column])
```

```python
column_correlation = df.corr().T
column_correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


df_scaled = column_correlation[['Age', 'Education_Num', 'Marital_Status','Relationship','Capital_Gain','Hours_Per_Week', 'Income_Bracket']]
fig, ax = plt.subplots(figsize=(8,4))
sns.heatmap(df_scaled.corr(),annot=True,linewidths=.5)
plt.show()
```

```python
X = df.loc[:,['Age', 'Education_Num', 'Marital_Status','Relationship','Capital_Gain','Hours_Per_Week']]            # storing into x and y
y = df['Income_Bracket']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler  
scaler = MinMaxScaler() #standard
X_train = scaler.fit_transform(X_train)
#X_train=X_train(scale)
X_train

X_test = scaler.fit_transform(X_test)
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score 
```

```python
inc = pd.DataFrame({"Actual":y_test, "Predicted": y_pred})
inc
```

```python
print("Accuracy", accuracy_score(y_pred, y_test))
```
