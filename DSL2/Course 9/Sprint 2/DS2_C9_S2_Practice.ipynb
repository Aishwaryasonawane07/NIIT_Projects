```python
import numpy as np     # for numerical operation
import pandas as pd    # for dataframe
import matplotlib.pyplot as plt


from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
car_df = pd.read_csv("D:\Data Science\Course 9\DS2_C5_S2_Mtcars_Data_Practice.csv")
car_df.head()
```

```python
fig, ax  = plt.subplots()
ax.set_xlabel('xlabel', fontsize=12)
ax.set_ylabel('ylabel', fontsize=12)
car_df.plot.scatter(x = 'wt', y = 'mpg', s = 20, ax=ax);
plt.show()
```

# Task 2

```python
X = np.array(car_df["wt"]).reshape((-1, 1))
y = np.array(car_df["mpg"])
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=200)
```

```python
s_model = LinearRegression().fit(X_train, y_train) # y = a +bx
s_r_sq = s_model.score(X_train, y_train)
```

```python
print('coefficient of determination:', s_r_sq)
print('intercept:', s_model.intercept_)
print('slope:', s_model.coef_)
```

```python
y_pred = s_model.predict(X_test)
print('predicted response:', y_pred, sep='\n')
```

```python
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
RMSE = mean_squared_error(y_test, y_pred, squared=False)
print(MSE, MAE, RMSE)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x_label", fontsize=12)
ax.set_ylabel("y_label", fontsize=12)
plt.ylim(0,30)

car_df.plot.scatter(x="wt", y = "mpg", s=30, ax=ax)
plt.plot(X_test, y_pred, color="blue", linewidth=2)
plt.show()
```
