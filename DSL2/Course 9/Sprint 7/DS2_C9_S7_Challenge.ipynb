```python
for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        # Calculate mean of the column
        mean = df[column].mean()
        print(f"Mean of {column}: {mean}")
        # Fill null values with the mean
        df[column].fillna(mean, inplace=True)
    elif df[column].dtype == 'object':
        # Calculate mode of the column
        mode = df[column].mode().values[0]
        print(f"Mode of {column}: {mode}")
        # Fill null values with the mode
        df[column].fillna(mode, inplace=True)
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = le.fit_transform(df[column])
```

```python
# define lists to gather results for plotting later
results, names = list(), list()
```

```python
from sklearn.model_selection import train_test_split
X = df.loc[:,(df.columns!="price")]
y = df.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8, random_state=5)
```

# Baseline : Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import RepeatedKFold,cross_val_score

# Initialize linear regression model
reg = LinearRegression()

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits= 10, n_repeats=3, random_state = 42)

n_score = cross_val_score(reg, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1, error_score='raise')
#(reg, X, y, scoring=,cv=cv,n_jobs=-1)

names.append('Linear')
results.append(np.mean(np.sqrt(n_score)))

print("MSC : %.3f (%.3f)" %(np.mean(n_score), np.std(n_score)))
```

# Bagging with Random Forest (For comparison)

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# define the model
model = RandomForestRegressor()

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats = 3, random_state= 42)
n_score = cross_val_score(model,X, y, scoring ='neg_mean_squared_error' , cv=cv, n_jobs=-1)

results.append(n_score)
names.append('RF')
# report performance

print("MSE score %.3f (%.3f)" %(np.mean(n_score), np.std(n_score)))
```

# Boosting with Adaboost

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
# define the model
model = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=7),n_estimators=50)

# evaluate the model
cv= RepeatedStratifiedKFold(n_splits=10, n_repeats = 3, random_state = 42)
n_score = cross_val_score(model , X , y, scoring ='neg_mean_squared_error' , cv=cv, n_jobs=-1)

results.append(n_score)
names.append("ADA")

# report performance
print("MSE %.3f(%.3f)" % (np.mean(n_score), np.std(n_score) ))
```

# Stacking

```python
# required Python libraries
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import StackingRegressor
```

```python
def get_stacking():
    model1=[]
    model1.append(("Linear",LinearRegression()))
    model1.append(("Decision",DecisionTreeRegressor(max_depth=7)))
    model1.append(("rf_Regressor",RandomForestRegressor(max_depth=7)))
    #model1.append(("SVR",SVR()))
    level1 = LinearRegression()
    model = StackingRegressor(estimators=model1, final_estimator=level1, cv=3)
    return model

stack_model=get_stacking()
score = cross_val_score(stack_model,X,y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1, error_score='raise')
results.append(np.mean(np.sqrt(score)))
names.append('stacking')
```
