```python
# import 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import mean
from numpy import std
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("D:\Data Science\Course 9\DS2_C5_S6_AllRent_Data_Challenge.csv")
```

```python
df.sample(5)
```

```python

df = df.drop(columns = ["Unnamed: 0", "photos","manager_id" ])
```

```python
df.isnull().sum()
```

```python
for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        # Calculate mean of the column
        mean = df[column].mean()
        print(f"Mean of {column}: {mean}")
        # Fill null values with the mean
        df[column].fillna(mean, inplace=True)
    elif df[column].dtype == 'object':
        # Calculate mode of the column
        mode = df[column].mode().values[0]
        print(f"Mode of {column}: {mode}")
        # Fill null values with the mode
        df[column].fillna(mode, inplace=True)

# Verify if all the null values have been filled
print(df.isnull().sum().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = le.fit_transform(df[column])
```

```python
from sklearn.model_selection import train_test_split
X = df.loc[:,(df.columns!="price")] 
y = df.loc[:,(df.columns == "price")]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=20)
```

```python
# evaluation imports
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import mean_absolute_error 
```

```python
from sklearn.tree import DecisionTreeRegressor

# define the model
model = DecisionTreeRegressor()

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1, error_score='raise')
#n_scores = cross_val_score()
                           
# report performance
print('neg_mean_squared_error: ' % (mean(n_scores), std(n_scores)))
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score
from sklearn.model_selection import RepeatedKFold,cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(max_depth=6)

rf_model.fit(X_train,y_train)

y_pred=rf_model.predict(X_test)

print("accuracy score of model : ",accuracy_score(y_pred,y_test))
print("Precision score of model : ",precision_score(y_pred,y_test,average="weighted"))
print("F1_score of model :",f1_score(y_pred,y_test,average="weighted"))
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=6)
model.fit(X_train,y_train)
ypred=model.predict(X_test)
print("accuracy score of model : ",accuracy_score(ypred,y_test))
print("Precision score of model : ",precision_score(ypred,y_test,average="weighted"))
print("F1_score of model :",f1_score(ypred,y_test,average="weighted"))
```
