```python
column_correlation = df.corr()
column_correlation
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(column_correlation,annot=True,linewidths=.5, ax=ax)

plt.show()
```

```python
feature_corr=column_correlation.corr()[["price"]].sort_values("price",ascending=False)[1:]
```

```python
X=df.loc[:,feature_corr.index[:4]]# Top 3 correlations
y=df.price
```

Train test split and scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=13)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse

model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)

print("Training score of Model : ",model.score(X_train,y_train))
print("Root mean squared error of model :",mse(y_pred,y_test,squared=False))
```
