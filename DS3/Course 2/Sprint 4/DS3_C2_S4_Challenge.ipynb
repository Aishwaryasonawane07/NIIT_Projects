# Task 2 :
Represent speech text in vector form using BOW. Supply the following for the words 'dedicated'and 'nation.
- a) Bow vector representation.
- b) Dictionary values

```python
clean_text = clean_text.lower().replace('.', '').replace(',', '').replace('\n', '')
```

```python
import collections
from typing import Dict,List, Tuple
```

```python
def word2bow(words : List[str], dictinory : Dict[str,int])->List[Tuple[int,int]]:
    word_frequencies = collections.defaultdict(int)
    for word in words:
        if word not in dictionary:
            dictionary[word] = len(dictionary)
        word_frequencies[dictionary[word]] +=1
        
    return list(word_frequencies.items())

dictionary = {}

print('\n Bag of Words representation : \n', word2bow(clean_text.split(),dictionary))
print('Dictionary : \n', dictionary)
```

```python
dedicated_dict_value = dictionary['dedicated']
nation_dict_value = dictionary['nation']


print('Dictionary value for "dedicated":', dedicated_dict_value)
print('Dictionary value for "nation":', nation_dict_value)
```

```python
doc = nlp('dedicated')
print('Word vector representation :',doc.vector)
```

```python
doc = nlp('nation')
print('Word vector representation :',doc.vector)
```

# Task 3 : 
For the speech texts, how will you define patterns and implement token-based matching to obtain the following outcomes?
- a) battlefield
- b) battle-field


```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
pattern = [ {'LOWER': 'battle'},{'IS_PUNCT':True},{'LOWER':'field'}]
```

```python
matcher.add("battlefield",[pattern])
doc = nlp(file)
matches = matcher(doc)
```

```python
for match_id,start,end in matches:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start:end]
    print(span.text)
```

```python
from spacy.matcher import PhraseMatcher

matcher=PhraseMatcher(nlp.vocab)
patterns=[nlp("battlefield"),nlp("battle-field")]
matcher.add("battlefield",patterns)
doc=nlp("battlefield ! battle-field!")
matches=matcher(doc)

for match_id,start,end in matches:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start:end]
    print(span.text)
```

# Task 5 :
For the speech texts, the phrases 'long endure' and 'resting place' should be matched using defined patterns ['long ebdure', 'resting place'].
- How will you set the attributes to achieve this ?

```python
matcher=PhraseMatcher(nlp.vocab,attr = "LOWER")
terms=["long endure",'resting place']
patterns=[nlp.make_doc(text) for text in terms] 

matcher.add("Name", patterns)
doc=nlp(clean_text)
matches=matcher(doc)
for match_id,start,end in matches:                                                  
    span = doc[start:end]
    print(span.text)                                                                  
```
