```python
doc1 = nlp(file1)
```

```python
def count_sentiment_words(text):
    docs=nlp(text)
    text=" ".join(str(doc) for doc in docs if(doc.is_stop==False))
    good=['good', 'great', 'excellent', 'positive', 'wonderful', 'amazing', 'fantastic', 'terrific', 'superb', 'outstanding']
    bad=['bad', 'terrible', 'awful', 'horrible', 'negative', 'sad', 'depressing', 'unhappy', 'gloomy', 'bleak']
    neutral=['neutral', 'okay', 'average', 'ordinary', 'typical', 'commonplace', 'mundane', 'routine', 'standard', 'usual']
    good_count=sum([text.lower().count(word) for word in good])
    bad_count=sum([text.lower().count(word) for word in bad])
    neutral_count=sum([text.lower().count(word) for word in neutral])

    return {'good': good_count, 'bad': bad_count, 'neutral': neutral_count}
sentiments=count_sentiment_words(doc1)
print('The proportion of words in each sentiment : ')
print(sentiments)
```

```python
doc2 = nlp(file2)
```

```python
sentiments=count_sentiment_words(doc2)
print('The proportion of words in each sentiment : ')
print(sentiments)
```

```python
doc3 = nlp(file3)
```

```python
sentiments=count_sentiment_words(doc3)
print('The proportion of words in each sentiment : ')
print(sentiments)
```
