# Task -2
Clean the file called "challenge_train.txt"

```python
import re
from nltk.corpus import stopwords
import nltk
import string
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
file=open('/content/challenge_train.txt','r').read()
```

```python
def cleaning(df):
  df = df.lower()
  # Remove special characters
  special_chars = ["!",",",".","~",":","-","?"]
  for char in special_chars:
    df = df.replace(char, "")

  # Tokenize the text into words
  words = nltk.word_tokenize(df)

  # Remove stop words
  stop_words = set(stopwords.words('english'))
  words = [word for word in words if not word in stop_words]

  # Join the words back into a string
  cleaned_text = ' '.join(words)

  return cleaned_text
```

```python
cleaned_text_train = cleaning(df=challenge_train)
cleaned_text_train
```

# Task -6
Test the model with various k-values and threshold
levels. Display the performance of each in a table.

```python
import numpy as np
import pandas as pd

check = []
for k in np.arange(1,10,1):
  for t in np.arange(0.1,1,0.1):
    test_num, precision, recall = model.test('cleaned_test_df.txt', k=k, threshold=t)
    f_score = (2*precision*recall)/(precision+recall)
    check.append(f_score)

list1 = []
list2 = []

for k in np.arange(1,10,1):
  for t in np.arange(0.1,1,0.1):
    list1.append(k) 
    list2.append(t)
    
mydf = pd.DataFrame([list1,list2,check]).transpose()
mydf.columns = ['K Value','Threshold value','f_score']
mydf.sort_values('f_score',ascending=False)
mydf
```
